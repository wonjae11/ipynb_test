```python
from bs4 import BeautifulSoup as bs

html_str = '''
<html>
    <body>
        <div id="content">
            <ul class = 'industry'>
                <li>인공지능</li>
                <li>빅데이터</li>
                <li>스마트팩토리</li>
            </ul>
            <ul class = 'comlang'>
                <li>Python</li>
                <li>C++</li>
                <li>Javascript</li>
            </ul>
        </div>
    </body>
</html>
'''

soup = bs(html_str, "html.parser")
```

```python
first_ul = soup.find('ul')
print(first_ul)
print(first_ul.text)
```

```python
all_li = first_ul.findAll('li')
print(all_li)
print(all_li[0])
print(all_li[0].text)

for li in all_li:
    print(li.text)
```

```python
second_ul = soup.find('ul', class_='comlang')
print(second_ul)
print(second_ul.text)
```

```python
first_ul = soup.select_one('ul.industry')
print(first_ul)
print(first_ul.text)
```

```python
first_all_li = first_ul.select('ul.industry > li')
print(first_all_li)
print(first_all_li[1])
print(first_all_li[1].text)
```

```python
second_ul = soup.select('li')
print(second_ul)
print(second_ul[0].text)
```

```python
idid = soup.select('#content')
print(idid)
print(idid[0].text)
```

```python
query = input("검색어를 입력하세요>>>")
response = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query="+query)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
links = soup.select(".news_tit")
print(query)
for link in links:
    title = link.text
    url = link.attrs['href']
    print(title, url)
```
